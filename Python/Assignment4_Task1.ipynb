**Assignment 4, Task 1: Read a File and Handle Errors**

Problem Statement:  Write a Python program that:

1.   Opens and reads a text file named sample.txt.

2.   Prints its content line by line.

3.   Handles errors gracefully if the file does not exist.

```python
try:
    with open('Sample1.txt','rt') as file_handler: # open file in read only mode
        print("Reading file content:\n")
        eof = False # flag to detect end of file
        i=1
        while eof == False:
            line = file_handler.readline() # read contents of file line by line
            if line=="": #readline return empty space upon encountering end of file
                break
            else:
                print("line ",i,": ",line,end="",sep="")
                i+=1
                                  
except FileNotFoundError as e:    
    print("Error: The file ",str(e).split(":")[-1].strip()," was not found",sep="") # extract file name from error message
    # error message is typecasted into a string for extraction
        
except (PermissionError, UnicodeDecodeError, OSError) as e: #catch exception
    print(e) # print the exception message
```

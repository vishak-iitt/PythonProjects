# Regular Expressions

There are 12 punctuation characters that are used in regex : 

$ ( ) * + . ? [ \ ^ { |

These 12 punctuation characters that make regular expressions work are called meta-characters. 

If you want your regex to match them literally, you need to escape them by using the backslash (\) in front of them. 


The punctuation characters, namely, ],-,} are not included in this set.

] and - become meta-characters only after an unescaped [

} becomes a metacharacter only after an unescaped {

There is no need to ever escape }



The symbols \  ^  - ] are the only ones that have special function inside a character class. Rest other need not be escaped when written inside the character class.

**^**

Matches position at start of string.

With (?m), also matches immediately after each \n.


**$**

Matches position at end of string.

Also matches before a trailing newline.

With (?m), matches before every \n.


**\A**

Matches only at index 0.

Immune to all flags.


**\Z**

Matches only at the absolute end of the string

Does not match before a final trailing newline

Unaffected by (?m) and (?s)

```python
text = "alpha\nbeta\nomega"
```

```python
import re 

text = "alpha\nbeta\nomega"

print(re.findall(r"^beta", text))
# []

print(re.findall(r"(?m)^beta", text))
# ['beta']

print(re.findall(r"\Abeta", text))
# []

print(re.findall(r"(?m)\Abeta", text))
# []
```

```python
import re 

text = "\nalpha\nbeta\nomega"

print(re.findall(r"^alpha", text)) #[]

text = "\n alpha\nbeta\nomega"

print(re.findall(r"^alpha", text)) #[]

text = " \nalpha\nbeta\nomega" #[]

print(re.findall(r"^alpha", text))
```

```python
text = " alpha\nbeta\nomega" 

print(re.findall(r"^alpha", text)) #[]

print(re.findall(r"\Aalpha", text)) # []

text = "\nalpha\nbeta\nomega" 
print(re.findall(r"\Aalpha", text)) # []
```

```python
text = "\n alpha\nbeta\nomega"

print(re.findall(r"^alpha", text)) # []
```

```python
text = "alpha\nbeta\nomega"

print("1. ",re.findall(r"omega$", text))
# ['omega']

print("2. ",re.findall(r"omega\Z", text))
# ['omega']

text = "alpha\nbeta\nomega\n"
print("3a. ",re.findall(r"omega$", text))
# ['omega']

text = "alpha\nbeta\nomega\n"
print("3aa. ",re.findall(r"omega\Z", text))
# []
```

```python
text = "alpha\nbeta\nomega\n " # notice there is a space after the final \n
print("3b.",re.findall(r"omega$", text))
#[]

text = "alpha\nbeta\nomega\n "
print("3b.",re.findall(r"omega\Z", text))
#[]

text = "alpha\nbeta\nomega \n"
print("3c.",re.findall(r"omega$", text))
# []
```

```python
text = "alpha\nbeta\nomega\n"
print("4. ",re.findall(r"omega\Z", text))
# []

text = "xya\nb\nc"

print("5. ",re.findall(r".$", text))
# ['c']

print("6. ",re.findall(r"(?m).$", text))
# ['a', 'b', 'c']
```

```python
text = "xya\nb\nc"
print("7. ",re.findall(r"(?m).\\Z", "a\nb\nc"))
# ['c']

text = "a\nb\nc"

print("8. ",re.findall(r"a.*c", text))
# []

print("9. ",re.findall(r"(?s)a.*c", text))
# ['a\nb\nc']

```

```python
text = "a\nb\nc"

print("10. ",re.findall(r"(?ms)^b.*$", text))
# ['b\nc']

text = "abc\ndef"

print("11. ",re.findall(r"^def", text))
# []

print("12. ",re.findall(r"(?m)^def", text))
# ['def']
```

```python
print("13. ",re.findall(r"\Adef", text))
# []

text = "start\nmiddle\nend"

print("14. ",re.findall(r"\Astart.*end\Z", text, re.S))
# ['start\nmiddle\nend']

text = "start\nmiddle\nend"

print("15. ",re.findall(r"\Astart.*end\Z", text, re.S))
# ['start\nmiddle\nend']
```

```python
text = "x\ny\nz"

print("16. ",re.findall(r"(?m)^.", text))
# ['x', 'y', 'z']

print("17. ",re.findall(r"\A.", text))
# ['x']

text = "omega\n"

print("18. ",re.search(r"omega$", text)  ) # Match
```

```python
print("19. ",re.search(r"omega\Z", text))  # Match

print("20. ",re.search(r"\nomega\Z", text))
# None
```

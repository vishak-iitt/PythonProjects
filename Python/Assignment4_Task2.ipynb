**Assignment 4, Task 2: Write and Append Data to a File**
 
Problem Statement: Write a Python program that:
1.   Takes user input and writes it to a file named output.txt.
2.   Appends additional data to the same file.
3.   Reads and displays the final content of the file.

```python
try:
    with open('output.txt','wt') as file_handler: # open file in write mode
        st = input("Enter text to be written to the file ")
        file_handler.write(st+"\n") # write data to the file
    with open('output.txt','at') as file_handler: # open file in append mode
        st = input("Enter additional text to append ")
        file_handler.write(st+"\n")
    with open('output.txt','rt') as file_handler: # open file in read only mode
        data = file_handler.read()  # read from the file
        print("Final content of output.txt:\n",data,sep="") 
                                  
except FileNotFoundError as e:    
    print(f"Error: The file {e.filename} was not found")     
        
except (PermissionError, UnicodeDecodeError, OSError) as e: #catch exception
    print(e) # print the exception message
```

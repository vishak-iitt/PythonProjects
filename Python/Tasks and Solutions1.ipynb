Problem Statement: Write a Python program that:

# Number Guessing game

```python
import random
count = 0
target = random.randint(1,50)

while count<=10:
    ch = int(input("Guess a number between 1-50 "))
    count+=1
    if not(1<=ch<=50):
        continue
    elif ch != target and ch>target:
        print("Wrong number! Try lower")
        continue
    elif ch != target and ch<target:
        print("Wrong number! Try higher")
        continue
    else:
        print(f"Success ! Number found in {count} number of steps")
        break
```

**Assignment 3, Task 1: Calculate Factorial Using a Function**


Problem Statement: Write a Python program that:
1.   Defines a function named factorial that takes a number as an argument and calculates its factorial using a loop or recursion.
2.   Returns the calculated factorial.
3.   Calls the function with a sample number and prints the output.

# Using Recursion

```python
def factorial():
    n = int(input("Enter a number "))
    def factorial1(n):
        if n <0:
            return "Unable to find factorial for negative numbers"
        elif n ==0:
            return 1
        else:
            return n*factorial1(n-1)
    print(f"Factorial of {n} is ",factorial1(n))
```

```python
factorial()
```

# Using loop

```python
n = int(input("Enter a number "))

if n < 0 :
    print( "Unable to find factorial for negative numbers")
else:
    prod=1
    for i in range(n,1,-1):
        prod*=i
    print(f"Factorial of {n} is ",prod)
```

# Serialization

Serialization is the process of converting an in-memory object (e.g., a Python dictionary, list, class instance) into a linear byte sequence or text representation so it can be:

stored (disk, database)

transmitted (network)

reconstructed later

Deserialization is the inverse operation.

# Serialization using JSON

```python
import random
stu_data = dict()
for i in range(1,5):
    stu = dict()
    stud = 'student'+str(i)
    for j in range(1,4):
        x = random.randint(0,100)
        stu['subject'+str(j)]=x
    stu_data[stud]=stu
print(stu_data)
```

```python
import json
# dump all at the same time - json.load → expects one valid full JSON structure
with open("jsonfile.json",'wt') as file_handler:
    json.dump(stu_data,file_handler,indent=4)
with open("jsonfile.json",'rt') as file_handler:
    data = json.load(file_handler)

print(data)
```

# Serialization using Pickle

```python
# In Pickle, data can be dumped either all at once or one at a time.
# if it was dumped all at once, the de-serialization shall also be all at once
# if dumped one at a  time, de-serialization will also be one at a time.
```

Dump all data at the same time

```python
import pickle

with open("pfile.pkl",'wb') as file_handler:
    pickle.dump(stu_data,file_handler)
with open("pfile.pkl",'rb') as file_handler:
    data = pickle.load(file_handler)

print(data)
```

Dump data one at a time

```python
import pickle

with open("pfile.pkl",'wb') as file_handler:
    pickle.dump("",file_handler)

# dump one at a time
for x in stu_data.items():
    with open("pfile.pkl",'ab') as file_handler:
        pickle.dump(x,file_handler)

with open("pfile.pkl",'rb') as file_handler:
    while True:
        try:
            data = pickle.load(file_handler) # Deserealize one at a time
            print(data)
        except :
            break
```

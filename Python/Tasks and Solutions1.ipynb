**Assignment 1 , Task 1: Perform Basic Mathematical Operations**

Problem Statement: Write a Python program that does the following:

1.  Takes two numbers as input from the user.
2.  Performs the basic mathematical operations on these two numbers:

o	Addition

o	Subtraction

o	Multiplication

o	Division

3.  Displays the results of each operation on the screen.

```python
# input
n1 = float(input("Enter first the first number: "))
n2 = float(input("Enter first the second number: "))
```

```python
result_add = n1 + n2
result_subtract = n1 - n2
result_multiplication = n1 * n2
if n2!=0:
    result_division = n1/n2
else:
    result_division = "Not possible"
```

```python
print("Addition: ",result_add)
print("Subtraction: ",result_subtract)
print("Multiplication: ",result_multiplication)
print("Division: ",result_division)
```

**Assignment 2, Task 1 : Module 3: Control Structures in Python**

Check if a Number is Even or Odd

Problem Statement:  Write a Python program that:
1. 	Takes an integer input from the user.
2. 	Checks whether the number is even or odd using an if-else statement.
3. 	Displays the result accordingly.
Expected Output:

```python
n = int(input("Enter a number")) # Take input from user and typecast into integer
if n%2==0: # if remainder when divided by 2 is 0
    print(f"{n} is an even number")
else: # else if remainder is non zero
    print(f"{n} is an odd number")
```

**Assignment 2, Task 2: Sum of Integers from 1 to 50 Using a Loop**

Problem Statement: Write a Python program that:
1.   Uses a for loop to iterate over numbers from 1 to 50.
2.   Calculates the sum of all integers in this range.
3.   Displays the final sum.

```python
summ=0 # initialize the variable summ to 0
for i in range(1,51,1): # i varies from 1 to 51 ( exclusive) in steps of 1
    summ+=i # Add i to summ
print("The sum of numbers from 1 to 50 is ",summ) # print final sum
```
